# VIF Critic EMD Training Notebook

This notebook trains and evaluates the VIF Critic using **Squared Earth Mover Distance** loss (Hou et al. 2017).

**What is this model?**
- An MLP with classification output: 3 logits per dimension (one per class)
- Uses squared EMD loss = L2 distance between predicted and target CDFs
- Squared EMD penalizes predictions proportional to ordinal distance, with gradients proportional to error magnitude
- Input: State vector (text embeddings + time gaps + history EMA + user profile)
- Output: 10-dim alignment scores in {-1, 0, 1}

```python
# Setup
import os
import sys
from pathlib import Path

notebook_dir = Path.cwd()
if notebook_dir.name == "critic_training":
    os.chdir(notebook_dir.parent.parent)
sys.path.insert(0, ".")

import numpy as np
import torch
import torch.nn as nn
from torch.optim import AdamW
from torch.optim.lr_scheduler import ReduceLROnPlateau
import polars as pl
import matplotlib.pyplot as plt

np.random.seed(42)
torch.manual_seed(42)

if torch.cuda.is_available():
    device = "cuda"
elif torch.backends.mps.is_available():
    device = "mps"
else:
    device = "cpu"

print(f"Working directory: {os.getcwd()}")
print(f"Using device: {device}")
```

## 1. Data Loading & Exploration

```python
from src.vif.dataset import load_all_data, split_by_persona, merge_labels_and_entries

labels_df, entries_df = load_all_data()

print(f"Labels: {labels_df.shape}")
print(f"Entries: {entries_df.shape}")
print(f"Unique personas: {labels_df.select('persona_id').n_unique()}")
```

```python
from src.models.judge import SCHWARTZ_VALUE_ORDER

merged_df = merge_labels_and_entries(labels_df, entries_df)

print("Alignment value distribution per dimension:")
print("=" * 50)
for dim in SCHWARTZ_VALUE_ORDER:
    col = f"alignment_{dim}"
    series = merged_df.select(col).to_series()
    counts = {v: (series == v).sum() for v in [-1, 0, 1]}
    print(f"{dim:15s}: -1={counts[-1]:3d}, 0={counts[0]:3d}, +1={counts[1]:3d}")
```

```python
train_df, val_df, test_df = split_by_persona(labels_df, entries_df, seed=42)

print(f"Train: {len(train_df)} entries ({train_df.select('persona_id').n_unique()} personas)")
print(f"Val:   {len(val_df)} entries ({val_df.select('persona_id').n_unique()} personas)")
print(f"Test:  {len(test_df)} entries ({test_df.select('persona_id').n_unique()} personas)")
```

## 2. Model Architecture

```python
from src.vif.encoders import SBERTEncoder
from src.vif.state_encoder import StateEncoder
from src.vif.critic_ordinal import CriticMLPEMD, emd_loss_multi

text_encoder = SBERTEncoder("all-MiniLM-L6-v2")
print(f"Encoder: {text_encoder.model_name}")
print(f"Embedding dim: {text_encoder.embedding_dim}")

state_encoder = StateEncoder(text_encoder, window_size=3, ema_alpha=0.3)
print(f"State dim: {state_encoder.state_dim}")
```

```python
# Create model (EMD: output 30 logits = 10 dims × 3 classes)
model = CriticMLPEMD(
    input_dim=state_encoder.state_dim,
    hidden_dim=256,
    dropout=0.2,
)
model.to(device)

n_params = sum(p.numel() for p in model.parameters())
print(f"Model parameters: {n_params:,}")
print(model)
```

## 3. Training Loop

```python
from src.vif.dataset import VIFDataset
from torch.utils.data import DataLoader

print("Creating datasets (this caches embeddings)...")
train_dataset = VIFDataset(train_df, state_encoder, cache_embeddings=True)
val_dataset = VIFDataset(val_df, state_encoder, cache_embeddings=True)
test_dataset = VIFDataset(test_df, state_encoder, cache_embeddings=True)

batch_size = 16
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"Train batches: {len(train_loader)}")
print(f"Val batches: {len(val_loader)}")
print(f"Test batches: {len(test_loader)}")
```

```python
# Training setup (EMD loss)
optimizer = AdamW(model.parameters(), lr=0.001, weight_decay=0.01)
scheduler = ReduceLROnPlateau(
    optimizer, mode="min", factor=0.5, patience=10, min_lr=1e-5
)

history = {"train_loss": [], "val_loss": [], "lr": []}
checkpoint_path = "models/vif/notebook_emd_best.pt"
```

```python
# Training loop
n_epochs = 100
best_val_loss = float("inf")
patience_counter = 0
early_stop_patience = 20

print("Training...")
for epoch in range(n_epochs):
    model.train()
    train_loss = 0.0
    for batch_x, batch_y in train_loader:
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)

        optimizer.zero_grad()
        logits = model(batch_x)
        loss = emd_loss_multi(logits, batch_y)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
    train_loss /= len(train_loader)

    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for batch_x, batch_y in val_loader:
            batch_x, batch_y = batch_x.to(device), batch_y.to(device)
            logits = model(batch_x)
            val_loss += emd_loss_multi(logits, batch_y).item()
    val_loss /= len(val_loader)

    scheduler.step(val_loss)
    current_lr = optimizer.param_groups[0]["lr"]

    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)
    history["lr"].append(current_lr)

    if val_loss < best_val_loss - 0.001:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save(model.state_dict(), checkpoint_path)
        print(f"Epoch {epoch + 1:3d}: train={train_loss:.4f}, val={val_loss:.4f}, lr={current_lr:.6f} [BEST]")
    else:
        patience_counter += 1
        if epoch % 10 == 0:
            print(f"Epoch {epoch + 1:3d}: train={train_loss:.4f}, val={val_loss:.4f}, lr={current_lr:.6f}")

    if patience_counter >= early_stop_patience:
        print(f"Early stopping at epoch {epoch + 1}")
        break

print(f"\nBest validation loss: {best_val_loss:.4f}")
```

```python
# Plot training curves
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(history["train_loss"], label="Train")
axes[0].plot(history["val_loss"], label="Validation")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("EMD Loss")
axes[0].set_title("Training & Validation Loss")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(history["lr"])
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Learning Rate")
axes[1].set_title("Learning Rate Schedule")
axes[1].set_yscale("log")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

The model is collapsing to predicting all class-0 (neutral). Look at the training log — it tells the whole story:


Epoch   1: train=0.3937, val=0.2879 [BEST]
Epoch   2: train=0.3242, val=0.2864 [BEST]
Epoch  11: train=0.3231, val=0.2857
Epoch  21: train=0.3228, val=0.2856
Early stopping at epoch 22
The loss flatlines almost immediately. Let's verify: if the model predicts all class-0 for every sample, the expected squared EMD loss would be:

True class -1, predicting 0: pred CDF = [0, 1], target CDF = [1, 1] → EMD² = (1-0)² = 1.0
True class 0, predicting 0: EMD² = 0.0
True class +1, predicting 0: pred CDF = [0, 1], target CDF = [0, 0] → EMD² = (0-1)² = 1.0
Expected loss = fraction_non_zero × 1.0 ≈ 0.30, which exactly matches the training loss of ~0.32.

Why does this happen with squared EMD but not with CORAL or Soft Ordinal?

The root cause is gradient strength vs class imbalance. Your data is ~70% class-0 on average. The three losses handle this very differently:

Cross-entropy-based losses (CORAL, Soft Ordinal, CORN): The log in -log(p) goes to infinity as the predicted probability of the true class approaches zero. This creates very strong gradients for misclassified minority samples, enough to overcome class imbalance.

Squared EMD: The gradient is 2(CDF_pred - CDF_target), which is bounded in [−2, 2]. For a minority sample that's misclassified, the gradient is the same fixed magnitude regardless of how wrong the prediction is. With only ~30% of samples being non-zero, these bounded gradients from the minority are overwhelmed by the majority-class signal that says "predicting 0 is free."

The L1→L2 fix solved the gradient scaling problem (gradients now scale with error magnitude), but didn't solve the gradient magnitude problem. EMD gradients are fundamentally bounded, while cross-entropy gradients are unbounded — and that unbounded quality is exactly what you need to fight class imbalance.

This is why the CORAL and Soft Ordinal notebooks actually learned something useful (Spearman 0.28 and 0.39 respectively) while EMD collapsed to all-zeros despite having correct gradient direction.

## 6. Notes: Squared Earth Mover Distance

**How it works:**
- EMD measures the "work" needed to transform one distribution into another
- For discrete ordinal distributions, EMD can be computed as the distance between CDFs
- We use **squared** EMD (L2 between CDFs) per Hou et al. 2017: `EMD²(p, q) = sum_k (CDF_p(k) - CDF_q(k))²`

**Why squared (L2) instead of L1?**
- L1 EMD (`|CDF_p - CDF_q|`) gives gradients via `sign(x)` = always ±1, regardless of error magnitude
- This means an almost-correct prediction gets the same gradient as a maximally-wrong one
- The optimizer overshoots and the model can collapse to predicting a single class
- Squared EMD gives gradients proportional to `2(CDF_p - CDF_q)` — larger errors → larger gradients, smaller errors → gentler corrections

**Example:**
- If true class is -1 (CDF = [1, 1]) and predicted probs are [0.2, 0.5, 0.3]:
  - Pred CDF = [0.2, 0.7] (exclude last element, always 1.0)
  - Target CDF = [1.0, 1.0]
  - EMD² = (1.0 - 0.2)² + (1.0 - 0.7)² = 0.64 + 0.09 = 0.73

**Advantages:**
- Loss magnitude reflects ordinal distance between predictions and truth
- Smooth, well-scaled gradients (proportional to error, unlike L1)
- No hyperparameters beyond standard training settings
- Well-studied in optimal transport theory

**Trade-offs:**
- Requires softmax output (3 logits per dim, more params than K-1 approaches like CORN/CORAL)
- May converge slower than direct classification losses on small datasets
